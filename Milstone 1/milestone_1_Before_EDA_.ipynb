```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
```

```python
# Gender Distribution
plt.figure(figsize=(6,4))
data["Gender"].value_counts().plot(kind="bar", color="skyblue", edgecolor="black")
plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

```python
# Highest Qualification Count
plt.figure(figsize=(7,4))
data["Highest Qualification"].value_counts().plot(kind="bar", color="lightgreen", edgecolor="black")
plt.title("Highest Qualification Count")
plt.xlabel("Qualification")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

```python
# Top 10 Majors / Specializations
plt.figure(figsize=(8,4))
data["Major / Specialization"].value_counts().head(10).plot(kind="bar", color="salmon", edgecolor="black")
plt.title("Top 10 Majors / Specializations")
plt.xlabel("Major / Specialization")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

```python
# Average Years of Experience by Preferred Job Role
plt.figure(figsize=(9,4))
data.groupby("Preferred Job Role")["Years of Experience"].mean().sort_values().plot(kind="bar", color="orange", edgecolor="black")
plt.title("Average Years of Experience by Preferred Job Role")
plt.xlabel("Preferred Job Role")
plt.ylabel("Average Experience (Years)")
plt.tight_layout()
plt.show()
```

```python
# Top 10 Most Common Skills
skills_list = data["Skills"].dropna().str.split(",").sum()
skills_flat = [s.strip() for s in skills_list]
skill_counts = Counter(skills_flat)
top_skills = dict(skill_counts.most_common(10))

plt.figure(figsize=(8,4))
plt.bar(top_skills.keys(), top_skills.values(), color="plum", edgecolor="black")
plt.title("Top 10 Most Common Skills")
plt.xlabel("Skills")
plt.ylabel("Frequency")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Age Distribution
plt.figure(figsize=(6,4))
plt.hist(data["Age"], bins=8, color="lightcoral", edgecolor="black")
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Number of Candidates")
plt.tight_layout()
plt.show()
```
